```python
from google.colab import drive
drive.mount('/content/drive')

import matplotlib.pyplot as plt
import numpy as np

# we are using load_img available in keras directly
from tensorflow.keras.utils import load_img
from PIL.ImageOps import autocontrast
from skimage.util import img_as_int
from skimage.segmentation import relabel_sequential
from skimage.transform import rescale, resize, downscale_local_mean

from skimage.color import label2rgb
from skimage.morphology import label
from skimage.exposure import rescale_intensity
from skimage import io

from distinctipy import distinctipy
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import pandas as pd
root_path = r'/content/drive/MyDrive/CV/Lab10/TrayDataset'
label_names = pd.read_csv(root_path + '/classes.csv')

# list of different colors, colors the same as classes, each class will have a unique color
colors = distinctipy.get_colors(43, rng = 12)

# function that given an image path will plot the image. If a mask path is given as well, it will be printed
# if a model is given, it will predict the image with the given model and print it
def plot_segmentation(img_path, mask_path = None, model = None, reshape = False, path_internet = False,
                      figsize = [16, 4], cmap = plt.rcParams["image.cmap"]):
  plt.rcParams['figure.figsize'] = figsize

  # how many images are we plotting
  n = 1
  n += 2 if mask_path else 0
  n += 2 if model else 0
  fig, axes = plt.subplots(1, n)
  
  # read the image
  img  = io.imread(img_path)

  # resize the image to plot them with the same size
  if reshape:
    # resize the image to (160, 160)
    img = resize(img, img_size, anti_aliasing = True)

  # show original image
  axes[0].imshow(img)
  axes[0].set_title('Input Image')
  
  i = 1

  # we plot the target image if the image is not from the internet
  if not path_internet:
    labels = io.imread(mask_path)[:,:,0] # we read image and only one channel, this is already the labels
    if reshape:
      # we reshape, since it is label data, we do not use anti aliasing and use nearest neighbors to avoid correctly 
      # resize the image
      labels = resize(labels, img_size, anti_aliasing = False, order = 0, preserve_range = True)

    colors_target = [colors[int(lab) - 1] for lab in np.unique(labels)[1:]]

    # plot the label image
    axes[1].imshow(label2rgb(labels, bg_label = 0, colors = colors_target))
    axes[1].set_title('Target Segmentation Mask')

    # target overlay on image
    axes[2].imshow(label2rgb(labels, img, bg_label = 0, colors = colors_target))
    axes[2].set_title('Target Overlay')

    # plot a legend on the bottom
    patches = [ mpatches.Patch(color=colors[int(lab) - 1], label = label_names['_name'][lab] ) for idx, lab in enumerate(np.unique(labels)[1:]) ]
    fig.legend(title = 'Target', handles=patches, loc='lower center', borderaxespad=0., 
                   ncol = len(np.unique(labels)) - 1, 
                   prop={'size': 10} )
    i += 2

  # if there is a model
  if model:
    if path_internet:
      mask_path = img_path
    # construct an image Generator
    gen_img = ImageGenerator(1, img_size, [img_path], [mask_path])

    # predict the image
    predicted_labels = model.predict(gen_img)[0]
    labels = np.argmax(predicted_labels, axis=-1)

    colors_predicted = [colors[int(lab) - 1] for lab in np.unique(labels)[1:]]


    # plot the predicted labels
    axes[i].imshow(label2rgb(labels, bg_label = 0, colors = colors_predicted))
    axes[i].set_title('Predicted Labels')
    
    i += 1
    
    # predicted labels overlayed on image
    axes[i].imshow(label2rgb(labels, image = resize(img, labels.shape), bg_label = 0, colors = colors_predicted))
    axes[i].set_title('Predicted Labels Overlayed')

    # plot a legend, depending on what we have plotted before we plot the legend in one positon or another
    patches_pred = [ mpatches.Patch(color=colors[int(lab) - 1], label = label_names['_name'][lab] ) for idx, lab in enumerate(np.unique(labels)[1:]) ]
    if i >= 3:
      fig.legend(title = 'Predicted', handles=patches_pred, loc='lower center', borderaxespad=0., 
                   ncol = len(np.unique(labels)) - 1, bbox_to_anchor=(0.495, 0.10),
                   prop={'size': 10} )
    else:
      fig.legend(title = 'Predicted', handles=patches_pred, borderaxespad=0., frameon = False,
              prop={'size': 10}, loc='center left', bbox_to_anchor=(1, 0.5) )

  # remove ticks
  for ax in axes:
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

  plt.tight_layout()
  plt.show()
```

# Image Generator

*   Create an image generator class (similar to the tutorial) to iterate over the images and return a tuple corresponding to the batch number.
*   The generator should inherit from the Sequence class.
*   It should have \_\_init\_\_(), \_\_len\_\_() and \_\_getitem\_\_() methods.
*   The batch size, image size, image paths and mask paths should be initialized using the \_\_init\_\_ method.

```python
# solution
from tensorflow import keras
import numpy as np
from tensorflow.keras.utils import load_img


class ImageGenerator(keras.utils.Sequence):
    """Helper to iterate over the data (as Numpy arrays)."""

    def __init__(self, batch_size, img_size, img_paths, mask_paths):
      # batch size
      self.batch_size = batch_size
      # image size
      self.img_size = img_size
      # image paths
      self.input_img_paths = img_paths
      # mask path
      self.target_img_paths = mask_paths
      # amount of images
      self.n_images = len(mask_paths)

    def __len__(self):
      # denotes the number of batches per epoch
      return self.n_images // self.batch_size

    def __getitem__(self, idx):
      """Returns tuple (input, target) correspond to batch #idx."""
      i = idx * self.batch_size
      batch_input_img_paths = self.input_img_paths[i : i + self.batch_size]
      batch_target_img_paths = self.target_img_paths[i : i + self.batch_size]
      x = np.zeros((self.batch_size,) + self.img_size + (3,), dtype="float32")
      for j, path in enumerate(batch_input_img_paths):
          img = load_img(path, target_size=self.img_size)
          x[j] = img
      y = np.zeros((self.batch_size,) + self.img_size + (1,), dtype="uint8")
      for j, path in enumerate(batch_target_img_paths):
          img = load_img(path, target_size=self.img_size, color_mode="grayscale")
          y[j] = np.expand_dims(img, 2)
      return x, y
```

# U-Net Xception-style model

*   We will use the same model architecture that we followed in the tutorial for this exercise

```python
from tensorflow.keras import layers


def get_model(img_size, num_classes):
    inputs = keras.Input(shape=img_size + (3,))

    ### [First half of the network: downsampling inputs] ###

    # Entry block
    x = layers.Conv2D(32, 3, strides=2, padding="same")(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    previous_block_activation = x  # Set aside residual

    # Blocks 1, 2, 3 are identical apart from the feature depth.
    for filters in [64, 128, 256]:
        x = layers.Activation("relu")(x)
        x = layers.SeparableConv2D(filters, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)

        x = layers.Activation("relu")(x)
        x = layers.SeparableConv2D(filters, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)

        x = layers.MaxPooling2D(3, strides=2, padding="same")(x)

        # Project residual
        residual = layers.Conv2D(filters, 1, strides=2, padding="same")(
            previous_block_activation
        )
        x = layers.add([x, residual])  # Add back residual
        previous_block_activation = x  # Set aside next residual

    ### [Second half of the network: upsampling inputs] ###

    for filters in [256, 128, 64, 32]:
        x = layers.Activation("relu")(x)
        x = layers.Conv2DTranspose(filters, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)

        x = layers.Activation("relu")(x)
        x = layers.Conv2DTranspose(filters, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)

        x = layers.UpSampling2D(2)(x)

        # Project residual
        residual = layers.UpSampling2D(2)(previous_block_activation)
        residual = layers.Conv2D(filters, 1, padding="same")(residual)
        x = layers.add([x, residual])  # Add back residual
        previous_block_activation = x  # Set aside next residual

    # Add a per-pixel classification layer
    outputs = layers.Conv2D(num_classes, 3, activation="softmax", padding="same")(x)

    # Define the model
    model = keras.Model(inputs, outputs)
    return model
    
```

*   What do you understand from the 'average' parameter of the Jaccard similarity score function?
*   What are the different values of 'average' parameter? 
*   Which parameter is used for what type of problem?

---

**Solution**

Jaccard similarity is the size of the intersection divided by the size of the union of two sets. It is used to compare a set of predicted labels ($\hat y$) and their corresponding ground truth labels ($y$).

$$J(y, \hat{y}) = \frac{|y \cap \hat{y}|}{|y \cup \hat{y}|}$$

Given the following matrix, representing a binary attribute:
$$
\left(
\begin{matrix}
M_{00} & M_{10} \\
M_{01} & M_{11} \\
\end{matrix}
\right)
$$

where:
- $M_{11}$ represents the total number of attributes where A and B both have a value of 1.
- $M_{01}$ represents the total number of attributes where the attribute of A is 0 and the attribute of B is 1.
- $M_{10}$ represents the total number of attributes where the attribute of A is 1 and the attribute of B is 0.
- $M_{00}$ represents the total number of attributes where A and B both have a value of 0.

then sklearn computes the Jaccard similarity as:

$$J = \frac{M_{11}}{M_{11} + M_{10} + M_{01}} $$


- The `average` parameter specifies the type of averaging performed in the data. If none is selected, it will return the score for each class. Thus, for multiclass or multilabel problems, we need to specify how to aggregate the singular measures to have a global measure of goodness.

- It accepts 6 different. The following enumerate the different possible inputs, and type of problem it is used on:

- `None`: scores for each class are returned, it can be useful to have a specific understanding of how each class performs and know the strengths and weaknesses of the classifier.

- `binary`: only reports results for the class specified. Only works if the target label is binary. Thus, we should not use this in multiclass classification, unless we want to study one variable against the rest. But then it would not really be a multiclass problem.

- `micro`: calculates the metrics globally by counting the total true positives, false negatives and false positives. This is useful for 2D comparisons, for example, image similarity. It collects the decisions for all classes into a single matrix, and then computes the performance for each, taking into account the global.

- `macro`: calculate metrics for each label, and finds their unweighted mean. This is useful for multilabel case where we have binary label indicators. Basically, in macro, we compute the performance for each class, and then average over classes, unweighted. This is not great when we have a large imbalance of classes, since the average is easily biased by outliers. Since the values are pooled together, it can also give wrong measurements when there is a high class imbalance. This also works well to regularize the absense of some classes in the data.

 Over micro and macro, there is not a clear winner when there is class imbalance, since macro returns equal importance to each class, while micro gives equal importance to each sample, as can be seen in [source](https://www.cse.iitk.ac.in/users/purushot/papers/macrof1.pdf), thus, when there is a high imbalance, the following is preferred:

- `weighted`: calculates the metrics for each label, with their average and number of instances each class has (support). Similar to macro, but it will take into account class imbalance. This is useful for when there is a high imbalance of classes and can skew the result of the similarity. By averaging with the support of each class, we hope to find a more representative value.

- `samples`: calculates metrics for each instance, and finds their average, only useful in multilabel classification. It uses an array of samples to compute the weights of the average.

*   What are other evaluation metrics to compare the performance of segmentation?

---

**Solution**

The most common method for evaluating the effectiveness of a segmentation method is subjective evaluation, in which a human visually compares
the image segmentation results. Usually, we search for a few characteristics to deem a segmentation as good. First, region interiors should be simple and without holes. Second, boundaries should be simple and not jagged. Moreover, these should be spatially accurate. Third, regions should be uniform and regular with respect to the image. Finally, the adjacent regions need to be differentiated with respect to the current segmentation object.

Evaluating the segmentation result depends on if we have a certain set of labeled data or not. 

The following is a brief study of the different metrics and methods used on supervised tasks, since they are more widespread and easier to understand than the unsupervised ones.

Usually, while evaluating machine learning models, we classify our predictions in true and false positives and in true and false negatives. In image segmentation, however, it is not immediately clear what true positives are and how we can evaluate our predictions, so there are different methods in order to try to evaluate semantic and instance segmentation techniques. 


To do so, the most commonly metrics used are the following ones:

<ins>For semantic segmentation:</ins>

  + **Pixel accuracy**: compares each pixel with its own ground truth mask. However, this metric is problematic where there is a class imbalance since, for example, having a low percentage of "occupied" pixels means that just detecting the total image we can obtain a pretty high accuracy even without detecting nothing. 
  When working correctly, this metric is used for each separately class as well as globally across all classes evaluating something similar to a binary mask where:

  $$accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$

  being $TP$ the true positive values representing that a pixel has been correctly predicted to belong to the given class, $TN$ the true negatives representing that a pixel has been correctly identified to not belong to a given class, $FP$ the false positive where a pixel is said to belong to a class that it does not and $FN$ the false negatives where a pixel is said to not belong to a class that it does. 

  + **Intersection over Union ($I_{o}U$) also known as Jaccard's Index**: quantifies the percentage of overlapping between the target mask and the predicted output, measuring the number of pixels that are common between the target and the prediction divided by the total number of pixels in both masks together. 

  $$I_{o}U = \frac{target \cap prediction} {target \cup prediction} = \frac{TP}{TP + FN + FP}$$

  This metric is similar to the *Dice coefficient* one (explained in the next bullet point).

  Regarding the intersection $(A \cap B)$ and the union $(A \cup B)$ in the previous formula, it is important to mention that the first one includes the pixels found in the prediction mask <ins>and</ins> the ground truth mask, whereas the second one includes all the pixels in either both masks, which means the prediction mask <ins>or</ins> the target one.

  + **Dice coefficient**: very similar to the *Jaccard's index* this metric double counts the intersection (TP) measuring the overlap between two masks. Dice coefficient is usually used instead Jacacrd's because the first one is easily differentiable but not the second one. 

  $$Dice\_coefficient = \frac{2 \times intersection}{union + intersection} = \frac{2TP}{2TP + FN + FP}$$

<ins>For instance segmentation:</ins>

Whereas semantic segmentation models output a single segmentation mask (since they have a binary segmentation), instance segmentation ones are more complicated to evaluate since "*they are a collection of local segmentation masks describing each object detected in the image*" (cite: [source](https://www.jeremyjordan.me/evaluating-image-segmentation-models/)).
Because of the this, the evaluation methods in this case are pretty similar to the *semantic* ones, with the exception that the $I_{o}U$ metric now is calculated with respect to the masks and not to the bounding boxes. 

  + **Calculating precision**: in case of working with a collection of predicted masks instead of with a single one, we need to compare each predicted mask with each of the available target masks for the specific input. In this case, the $TP$ appear when a pair of target-prediction mask has an $I_{o}U$ score that exceeds a predefined threshold, the $FP$ when a predicted object mask has no relation with the ground truth object mask and a $FN$ when a ground truth object has no relation with a predicted object mask. Therefore, in this case, we work with <ins>precision</ins> and <ins>recall</ins> metrics.
    + <ins>Precision</ins>: describes the purity of the positive detections regarding the ground truth. 
    $$Precision = \frac{TP}{TP + FP}$$
    + <ins>Recall</ins>: describes the completeness of the positive predictions regarding the ground truth.
    $$Recall = \frac{TP}{TP + FN}$$

*   What are the challenges when you use a trained model in the wild?
*   How can you develop a model to be used in real world?

---

**Solution**

- The challenges are the same for any machine learning model. We need to generalize well the method to avoid overfitting and be able to predict future labels in different conditions. However, due to the nature of images, there are a few additional challenges:

  - Perspective/Rotation of the objects' axis. While translation can be "solved" with data augmentation techniques, rotation of the object on its axis can cause problems. For example, a car seen from the front is very different from the top.
  - Illumination variance. This is one of the reasons we think our "internet" images worked so poorly. In the original images, the contrast between the tray and the rest of the background was maximized. However, in our images, the contrast was not modified and thus, it had a harder time to segment the images.
  - Object occlusion. This can be "solved" with data augmentation techniques to simulate partial occlusions. However, it is more training needed to be done.
  - Image quality and resolution. While most neural nets do not work with large images, small images could cause failure due to the upsampling of the image.
  - Differences of the same object. For example, apples can come in various different colors, as well as many other things in the real world. Thus, if we only have green apples, the model will fail when we have a red apple if it only focuses on the color.
  - Sometimes, the model can have a high bias for a particular class. This results in the model's predictions to be skewed for that class. This can happen when we have unbalanced labels.
  - Obviously, images that are unseen in training the model will be impossible to predict in execution time. Thus, an updating mechanism should be given for the model to continuously learn from the data.

- Some ideas have already been presented in the previous bullet point, but the main idea is always to use more data. Since this is not trivial, as labeling images is a very time extensive process, there are some other solutions:
 - Data augmentation: rotation, translation, noise addition, downsampling, upsampling, partial occlusion of the current images. This will allow fixing many of the generalizations in the real world, however missing labels can still be a problem.
 - Add an unknown label threshold. This threshold will add a gate to the model, which basically will prevent the model from predicting unknown labels.

*   What is the difference between instance segmentation and semantic segmentation?

---

**Solution**

The objective in computer vision is not only to develop an algorithm capable of detecting images, but to obtain high level accuracy while processing the information. Image segmentation is a method where a digital image is "*broken down into various subgroups called Image segments, which helps in reducing the complexity of the image to make further processing or analysis of the image simpler*" (cite:[source](https://www.researchgate.net/publication/366310707_Image_Segmentation_using_Encoder-Decoder_model)). In order to accomplish this purpose, we want to assign labels to pixels using two types of segmentation: *Instance* and *semantic* one.

+ **Semantic segmentation**: this segmentation is the more general one. The objects shown in an image are grouped based on different categories, so all the objects that belong to the same class share the same label. Giving an example, referring to a street scene, semantic segmentation is able to discern the difference between a *pedestrian* and a *vehicle*, but not between the different types of *vehicles*.

  <img src="https://keymakr.com/blog/content/images/2021/05/-------------3-.jpg" alt="drawing" width="400"/>


+ **Instance segmentation**: this type of segmentation, however, is a refined version of the previous one. The objects shown in an image are now classified in different instances for each category. In this type of segmentation, each object receives a unique label, so it is usually implemented when the number of objects or their independence is relevant. Returning to the previous example, in this case, within each category of *pedestrian* or *vehicle*, each person would be differentiated, as well as whether the vehicle is a *car*, a *bus*, a *motorcycle*...

*   What is the difference between encoder and decoder? Should they be of the same type of neural network?

---

**Solution**

The aim of Image segmentation is to process just the important segments, converting an image into a "*collection of regions of pixels that are represented by a mask or a labeled image*" (cite: [source](https://https://www.researchgate.net/publication/366310707_Image_Segmentation_using_Encoder-Decoder_model))


To do so, one of the techniques used (usually in the Semantic Segmentation already explained in the previous exercise), are the **encoder** and **decoder** networks. The main differences between these two lies on the fact that **encoder** networks takes an input image and generates a high-dimensional feature vector, aggregating also features at multiple levels, and it is usually a pre-trained classification network (VGG, ResNet...). Regarding to **decoder** networks, however, they work backwards. They take a high-dimensional feature vector and generate a semantic segmentation mask. Contrary to **encoder** ones, **decoder** networks decode the features aggregated by encoder at multiple levels projecting the discriminative features (that means the ones with lower resolution) learned by the encoder onto the pixel space (with higher resolution) to get dense classification.

Adding a little more information about the previous stated, we could also point that **encoder** networks perform convolution with filter banks to produce a set of feature maps then batch normalized and applied an element-wise ReLU (rectified linear non-linearity max (0,x)). After that, it is important to perform a max-pooling (which is used to achieve translation invariance over small spatial shifts in the input image) with a 2 x 2 window and stride 2 (that means a non-overlapping window) and the output obtained is a sub-sampled by a 2 factor. Considering all this information, however, it is important to know that "*while several layers of max-pooling and sub-sampling can achieve more translation invariance for robust classification there is a loss of spatial resolution of the feature maps*" (that means a loss in the image's boundary detail, which is not beneficial for segmentation). To overcome this problem, it is necessary to capture and store the boundary information in the encoder feature map before doing the sub-sampling.
Regarding the **decoder** network, it up-samples its input feature map using the memorized max-pooling indices from the corresponding to encode feature maps, producing a spare feature map that then in convolved with a trainable decoder filter bank to produce dense feature maps. In this kind of architecture it is important to mention that the decoder corresponding to the first encoder performs differently to the others since it produce a multichannel feature map (although its encoder input has 3 channels) unlike the others that produce features maps with the same number of size and channels as their encoder inputs.

(cite: [source](https://courses.cs.washington.edu/courses/cse576/17sp/notes/Sachin_Talk.pdf))


Semantic Segmentation architecture, can be though as an encoder network followed by a decoder one since not only requires the discriminations at pixel level but to project the discriminative features learned in different states of the encoder onto the pixel space ([source](https://medium.com/@fezancs/understanding-of-semantic-segmentation-how-segnet-model-work-to-perform-semantic-segmentation-5c426112e499)).


However, for a more extensive information about the join performance between *encoder* and *decoder* networks, we need to understood *SegNet Model* for *Semantic Segmentation*.

This model appeared due to the necessity to model appearance, shape and spatial-relationships (meaning the context) between different classes and has been designed as an efficient architecture for pixel-wise semantic segmentation using both *encode* and *decode* networks followed by a final pixelwise classification layer. The *encoder* network has 13 convolutional layers (corresponding to the firsts 13 convolutional layers in the VGG16 networks designed for object classification) and discard the fully connected layers in order to retain higher resolution feature maps at the deepest encoder output. This also reduces the number of parameters in this kind of architecture compare to others.

Regarding the *decode* layers, each *decode* layer is associated to a corresponding *encode* one, so there is no doubt that the *decode* network has 13 layers.

Finally, in order to produce class probabilities for each pixel independently, the final *decoder* output, is fed to a multi-class soft-max classifier.
So, answering the previous question, *SegNet* would be an example where *encode* and *decode* can perform together.

*   What is upsampling?

---

**Solution**

For a computer, all the information processed is in the form of numbers, so one of the big deals while processing images is the way to resize an image.

It is known as *upsampling* the increase of the number of rows and/or columns of the image. In other words, the increase of an image's dimensions. This technique can be performed in order to increase the image's quality or to "construct an image out of random vector sample mimicking an image from the ground-truth or real distribution*" ([source](https://medium.com/analytics-vidhya/downsampling-and-upsampling-of-images-demystifying-the-theory-4ca7e21db24a)) as happens with Generative Adversarial Network among others.

Contrary to the *downsampling*, where the idea of reducing the dimensionality of an image was pretty simple, with the *upsampling* we need to look for a way to fill the rows and columns in order to increase the dimensions. To do so, there are several options. One of them, could be, for example, filling the gaps just repeating the rows or columns in the original image. This technique would provoke that both the original and the new image were similar but not identical
. Another way, could be using advanced mathematical produces (as Nearest Neighbor Interpolation) by interpolating the new data between the rows or the columns. This is what we have done in resizing the target mask images for comparision for the Jaccard, otherwise the actual segments degenerated, and returned worse scores.

In order to upsample the resolution of a feature map exist different approaches like *unpooling* operations (such as Nearest Neighbor, Bed of Nails or Max Unpooling) which upsample the resolution by distributing a single value into a higher resolution. Nevertheless, transpose convolutions are by far the most popular ones (taking a single value from the low-resolution feature map and multiplying all the weight in the filter by this value, so it projects the weighted values into the output feature map). Other approaches would be adding skip connections (from earlier layers and summing the two feature maps, which should provide enough detail to reconstruct the shapes of segmentation boundaries) or the use of U-Net ("*a convolutional network modified in order to supplement a usual contracting network by successive layers, where pooling operators are replaced by upsampling operators to increase the output's resolution and using many feature channels to allow the network to propagate context information to higher resolution layers*") ([source](https://arxiv.org/pdf/1505.04597.pdf)).

[source](https://medium.com/analytics-vidhya/downsampling-and-upsampling-of-images-demystifying-the-theory-4ca7e21db24a) 
[source](https://www.jeremyjordan.me/semantic-segmentation/#upsampling)

*   What are feature channels in UNet? Why is this helpful in this architecture?

---

**Solution**

The U-Net architecture is based on a contracting path (which follows the typical convolutional network architecture) and an expansive path (where every step in the expansive path consisting on an upsampling of the
feature map followed by a 2x2 convolution).

The first one consists of a repeated application of 3x3 convolutions each of them followed by a ReLU and a 2x2 max pooling operation with stride 2 for downsampling. At each downsampling step, the number of feature channels is doubled.

Regarding the expansive path, considering the upsampling of the feature map followed by a 2x2 convolution that halves the number of feature channels we have also a concatenation with the correspondingly cropped
feature map from the contracting path (needed due to the loss of border pixels in every convolution), and two 3x3 convolutions, each followed by a ReLU. Finally, in the last layer, a 1x1 convolution is used to map each 64-component feature vector to the desired number of classes.

Considering all this information, we can conclude that the feature channels are nothing more than the number of channels at a given point in the network which allow propagating context information to higher resolution layers.
Consequently, the expansive path is more or less symmetric to the contracting path, and yields a U-shaped architecture.

It is also important to mention that UNet does not have any fully connected layers and only uses the valid part of each convolution for which the full context is available in the input image (allowing the seamless segmentation of arbitrarily large images by an overlap tile strategy, which helps the resolution to not be limited by the GPU memory while working with large images).

(All the previous information has been extracted from the official paper [U-Net: Convolutional Networks for Biomedical
Image Segmentation](https://arxiv.org/pdf/1505.04597.pdf))

*   What are the other popular semantic segmentation algorithms?
*   How are they different from each other?

---

**Solution**

Although there are a wide variety of algorithms in semantic segmentation, the three main ones are the **Fully Convolutional Network (FCN)**, the **U-Net** and the **Mask RCNN**.
However, since **U-Net** has already been explained in this section, we will compare the other two.

**Fully Convolutional Network (FCN)** uses various blocks of convolution and max pool layers to decompress an image to $1/32th$ of its original size. Then, at this level of granularity makes a class prediction and, at the end, using upsampling and deconvolution layers it resizes the image to its original dimensions. **Mask RCNN**, also known as **Faster R-CNN + FCN** is used for object detection and has 2 stages. During the first stage (Region Proposal Network (RPN)) it proposes candidate objects bounding boxes. On the second stage (Fast R-CNN) it extracts the features using RoIPool from each candidate box and performs classification and bounding-box regression. In order to make the inference faster, the features used in both stages can be shared.

Regarding more technical aspects, **FCN** typically do not have any fully connected layers (since they have been replaced by convolutional ones) and there are no limitations on image size, being the final image the same size as the original one. Moreover, it uses skip connections (connection that bypasses at least one layer) to recover the fine-grained spatial information lost in downsampling (capture semantic/contextual information). It is used to pass information from the downsampling step to the upsampling (recover spatial information) one merging features from various resolution levels since the network produces several feature maps with small sizes and dense representations, so it is necessary an upsampling to create an output with the same size as the input.

**RCNN**, however, has two outputs for each candidate object, a class label and a bounding-box offset. Then a third branch is added to output the object mask, which is binary and indicates the pixels where the object is in the bounding box. The additional mask output requires extraction of much finer spatial layout of an object (using a FCN) and it is distinct from the class and box outputs. Moreover, **RCNN** is more accurate than **FCN**.

*   What are the popular semantic segmentation datasets which are available for public access?


---

**Solution**

In order to test the different semantic segmentation algorithms, there are some datasets available for public access which use different metrics of evaluation:

+ **PASCAL VOC (Visual Object Classes)**: commonly used for object detection and segmentation it contains more than 11k images, for train and validation, and 10k images dedicated to the test. The segmentation challenge is evaluated using the mean Intersection over Union ($mI_{o}U$)
+ **PASCAL Context**: it is an extension of the 2010 PASCAL VOC dataset and contains around 10k images for training, 10k for validation and 10k for testing. The main difference, regarding the previous one, is the fact that the entire scene is segmented, providing more than 400 categories. The official evaluation metric, as in the previous case, is the $mI_{o}U$, although there are others published as the pixel Accuracy ($pixAcc$).
+ **COCO (Common Objects in COntext)**: in this case, there are two challenges for image segmentation: the *object detection* and the *stuff segmentation*.
The first one segments and categorizes objects into 80 categories, and it is composed of more than 200k images with over 500k object instance segmented (considering training, validation, test data for researchers and test data for the challenge). On the other hand, the second one uses data with large segmented parts of the images that contain almost the entire visual information. In this case, the metrics used are the Average Precision ($AP$) and the Average Recall ($AR$), but both of them use the $mI_{o}U$.
+ **Cityscapes**: it consists of a complex segmented urban scenes from 50 cities, composed of 23.5k images used for training and validation and 1.5k images used for testing. In this case, the images are fully segmented (something similar to the PASCAL Context case) with 29 classes (with 8 super categories: *flat*, *human*, *vehicle*, *construction*, *object*, *nature*, *sky*, *void*). Due to its complexity and its similarity with real urban scenes, this dataset is used to evaluate semantic segmentation models and autonomous driving applications. In this case, as well as in the previous cases, the performance is evaluated using the $mI_{o}U$.
